4.   Develop an Algorithm





```python
# Function to input monthly energy consumption data
def input_monthly_data(month):
    '''
    Function to input monthly energy consumption data from the user.
    Returns a dictionary containing the monthly consumption data.
    '''
    print(f"\nInput data for Month {month}:")
    electricity = float(input("Enter electricity usage (kWh): "))
    gas = float(input("Enter gas usage (cubic meters): "))
    water = float(input("Enter water usage (cubic meters): "))

    return {'Month': month, 'Electricity': electricity, 'Gas': gas, 'Water': water}

#Input monthlydata for all 12 months
monthly_data = [input_monthly_data(month) for month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']]

```

```python
import matplotlib.pyplot as plt

# Function to visualize monthly consumption trends
def monthly_consumption_trends(data):
    '''
    Function to visualize monthly consumption trends.
    '''
    months = [entry['Month'] for entry in data]
    electricity_usage = [entry['Electricity'] for entry in data]
    gas_usage = [entry['Gas'] for entry in data]
    water_usage = [entry['Water'] for entry in data]

    plt.plot(months, electricity_usage, label='Electricity')
    plt.plot(months, gas_usage, label='Gas')
    plt.plot(months, water_usage, label='Water')

    plt.xlabel('Month')
    plt.ylabel('Usage')
    plt.title('Monthly Consumption Trends')
    plt.legend()
    plt.show()
```

```python
# Function to compare energy types by consumption
def compare_energy_types(data):
    '''
    Function to compare energy types by consumption.
    '''
    energy_types = list(data[0].keys())[1:]  # Exclude 'Month' key
    total_usage = [sum(entry[type] for entry in data) for type in energy_types]
    plt.bar(energy_types, total_usage)
    plt.xlabel('Energy Type')
    plt.ylabel('Total Usage')
    plt.title('Comparison of Energy Types by Consumption')
    plt.show()

# Function to visualize annual total consumption breakdown by energy type
def annual_total_consumption(data):
    '''
    Function to visualize annual total consumption breakdown by energy type.
    '''
    energy_types = list(data[0].keys())[1:]  # Exclude 'Month' key
    total_usage = [sum(entry[type] for entry in data) for type in energy_types]
    plt.pie(total_usage, labels=energy_types, autopct='%1.1f%%')
    plt.title('Annual Total Consumption Breakdown by Energy Type')
    plt.show()


while True:
# Ask the user for analysis option
      print("\nAnalysis Options:")
      print("1. Monthly Consumption Trends")
      print("2. Compare Energy Types by Consumption")
      print("3. Annual Total Consumption Breakdown by Energy Type")
      print("4. Exit")
      choice = input("Enter your choice (1-4): ")

# Perform analysis based on user's choice
      if choice == '1':
          monthly_consumption_trends(monthly_data)
      elif choice == '2':
           compare_energy_types(monthly_data)
      elif choice == '3':
           annual_total_consumption(monthly_data)
      elif choice == '4':
           print("Exiting program.")
           break
      else:
           print("Invalid choice. Please enter a number from 1 to 4.")
```
